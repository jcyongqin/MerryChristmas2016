```python
import numpy as np
import scipy as sp
import scipy.misc


def f(x):
    return 2*x*x +3*x +1

print(sp.misc.derivative(f,2))
```

```python
import numpy as np
import scipy as sp
import scipy.integrate

f = lambda x:x**2

print(sp.integrate.quad(f,0,2))

print(sp.integrate.fixed_quad(f,0,2))
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def f(x):
    return [5*x[1] + 3,
           4*x[0] *x[0],
           x[1]*x[2]-1.5]

ans=sp.optimize.fsolve(f,[0,0,0])
print(ans)
print(f(ans))
```

```python
import numpy as np
import scipy as sp
import scipy.linalg
import matplotlib.pylab as plt

a=np.array([[1,3,5],[2,5,1],[2,3,8]])
b=np.array([10,8,3])

print(sp.linalg.solve(a,b))
```

```python
import numpy as np
import scipy as sp
import matplotlib.pylab as plt
import pylab
from mpl_toolkits.mplot3d import Axes3D



fig= pylab.figure()
ax = fig.add_subplot(111, projection='3d')


X = np.linspace(-10,10,30)
Y = np.linspace(-10,10,30)
X,Y=np.meshgrid(X,Y)
Z=(X*X)/3-(Y*Y)/3
ax.plot_surface(X,Y,Z,rstride=1,cstride=1)
plt.show()
```

```python
import numpy as np
import scipy as sp
import matplotlib.pylab as plt
import random


t =np.linspace(0, 2*np.pi, 1000, endpoint=True)
p =3*np.sin(3*t)
plt.subplot(111, polar=True)
color =plt.cm.Set2(random.choice(range(plt.cm.Set2.N)))
plt.plot(t, p, lw=3, color=color)
plt.grid(True)
plt.show()
```
